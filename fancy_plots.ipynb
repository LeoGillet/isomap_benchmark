```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
saved_folder = "saved/"
samples_subfolder = saved_folder+"samples/"
comp_subfolder = saved_folder+"components/"
ngbr_subfolder = saved_folder+"neighbors/"
```

```python
samples_runs = pd.read_csv(samples_subfolder+"run_results.csv", header=0, sep='\t')

x = range(len(set(samples_runs["N_SAMPLES"])))
minimum_values = list()
for i in range(0, len(samples_runs.get("SCORE_JACCARD"))-1, 5):
    mv = samples_runs.get("SCORE_JACCARD")[i:i+5]
    minimum_values.append(mv.mean())
```

```python
fig, ax = plt.subplots()
ax.bar(x, minimum_values, align='center')
ax.set_xticks(x)
_ = ax.set_xticklabels(set(samples_runs["N_SAMPLES"]))
print(set(samples_runs["N_SAMPLES"]))
plt.show()
```

```python
comp_runs = pd.read_csv(comp_subfolder+"run_results.csv", header=0, sep="\t")
```

```python
plt.figure()
plt.boxplot(comp_runs["N_COMPONENTS"], comp_runs["TIME_ISOMAP"])
plt.show()
```

```python
ngbr_runs = pd.read_csv(ngbr_subfolder + "run_results.csv", header=0, sep="\t")
x = range(len(set(ngbr_runs["N_NEIGHBORS"])))
minimum_values = list()
for i in range(0, len(ngbr_runs.get("TIME_ISOMAP"))-1, 5):
    mv = ngbr_runs.get("TIME_ISOMAP")[i:i+5]
    minimum_values.append(mv.min())
```

```python
fig, ax = plt.subplots()
ax.bar(x, minimum_values, align='center')
ax.set_xticks(x)
_ = ax.set_xticklabels(set(ngbr_runs["N_NEIGHBORS"]))
plt.show()
```
